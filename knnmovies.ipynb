```python
df = pd.read_csv('netflix_movies_detailed_up_to_2025.csv')
print(df.head())
```

```python
df = np.array(df)
X = df[:, :-1]
Y = df[:, -1]
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
print(X_train[:4])
print(X_train.dtype)
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train,Y_train)
```

```python
from sklearn.preprocessing import LabelEncoder

# ... (Your existing code for loading and preprocessing data) ...

# Create a LabelEncoder instance
le = LabelEncoder()

# Fit the encoder to the target variable (Y)
# Include both training and test data to ensure all possible labels are encoded
le.fit(np.concatenate((Y_train, Y_test)))

# Transform both the training and test target variables
Y_train_encoded = le.transform(Y_train)
Y_test_encoded = le.transform(Y_test)

# Train the model using the encoded target variable
clf = KNeighborsClassifier()
clf.fit(X_train, Y_train_encoded)  # Use encoded Y_train
```

```python
score = clf.score(X_train,Y_train)
# The model was trained on data with 17 features.
# Create a sample data point with 17 features for prediction.
# Replace placeholders with actual values.
sample_data_point = [[10, 14, 0, 0, 0, 0, 0, 8, 0, 0, 0, 3, 0, 0, 7, 0, 0]]
print(clf.predict(sample_data_point))
```

```python
predict = clf.predict(X_test)
print(accuracy_score(Y_test_encoded, predict))
```

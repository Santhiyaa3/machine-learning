```python
data = pd.read_csv("Employee.csv")
print(data.head())
plt.figure(figsize=(8,5))
```

```python
x_col = data.columns[data.columns.str.contains('category', case=False)][0] if any(data.columns.str.contains('category', case=False)) else data.columns[0]
y_col = data.columns[data.columns.str.contains('Value', case=False)][0] if any(data.columns.str.contains('Value', case=False)) else data.columns[1]
sns.boxplot(x=x_col, y=y_col, data=data,color='pink')
plt.title("Box Plot of Value by Category")
plt.show()
```

```python
ax = sns.scatterplot(x="Age", y="ExperienceInCurrentDomain", data=data)
plt.show()
```

```python
ax = sns.scatterplot(x="Age", y="ExperienceInCurrentDomain", hue="PaymentTier", data=data)
plt.show()
```

```python
ax = sns.histplot(data["Age"], kde=True,color='purple')
plt.show()
```

```python
ax = sns.boxplot(x="Education", y="Age", data=data,color='violet')
plt.xticks(rotation=45)
plt.show()
```

```python
ax = sns.boxplot(x="PaymentTier", y="Age", data=data,color='blue')
plt.show()
```

```python
ax = sns.countplot(x="Gender", data=data,color='yellow')
plt.show()
```

```python
ax = sns.countplot(x="EverBenched", data=data,color='green')
plt.show()
```

```python
ax = sns.scatterplot(x="Age", y="ExperienceInCurrentDomain", hue="LeaveOrNot", data=data,)
plt.show()
```
